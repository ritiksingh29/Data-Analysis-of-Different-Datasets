```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("company_sales_data.csv")
```

```python
data.head()
```

```python
plt.plot(data['month_number'],data['total_profit'],label= "profit")
# plt.title("Delhi Temperature")
plt.xlabel("Month Number") 
plt.ylabel("Total Profit")
plt.xticks(data['month_number'])
plt.legend(loc=4)
plt.show()
```

```python
plt.plot(data['month_number'],data['total_profit'], color = "r", marker = "o", linestyle= "--", linewidth = 3,label= "profit")
plt.xlabel("Month Number") 
plt.ylabel("Total Profit")
plt.xticks(data['month_number'])
plt.legend(loc=4)
plt.show()
```

```python
plt.plot(data['month_number'],data['facecream'], color = "r", marker = "o", linestyle= "--", linewidth = 3,label= "profit")
plt.plot(data['month_number'],data['facewash'], color = "g", marker = "o", linestyle= "--", linewidth = 6,label= "profit")
plt.plot(data['month_number'],data['toothpaste'], color = "y", marker = "o", linestyle= "--", linewidth = 3,label= "profit")
plt.plot(data['month_number'],data['bathingsoap'], color = "b", marker = "o", linestyle= "--", linewidth = 3,label= "profit")
plt.plot(data['month_number'],data['shampoo'], color = "c", marker = "o", linestyle= "--", linewidth = 3,label= "profit")
plt.plot(data['month_number'],data['moisturizer'], color = "m", marker = "o", linestyle= "--", linewidth = 2,label= "profit")
plt.xlabel("Month Number") 
plt.ylabel("Sold Unit")
plt.xticks(data['month_number'])
plt.legend(loc=4)
plt.show()
```

```python
x= data['month_number']
y= data['toothpaste']
plt.scatter(x,y)
plt.xlabel("Month Number") 
plt.ylabel("toothpaste")
plt.xticks(data['month_number'])
plt.legend(loc=4)
plt.show()
```

```python
width = 0.4
 
plt.bar(data['month_number'],data['facecream'], width , color = "b",
        label =" facecream") #visible=False
 
plt.bar(data['month_number']+ width,data['facewash'], width , color = "g",
        label =" Facewash") 
plt.xlabel("Month Number") 
plt.ylabel("Facewash & Facecream")
plt.xticks(data['month_number'])
plt.legend(loc=4)
plt.show()
```

```python
plt.bar(data['month_number'],data['bathingsoap'], color = "b",
        label ="bathing soap") 
plt.xlabel("Month Number") 
plt.ylabel("Facewash & Facecream")
plt.xticks(data['month_number'])
plt.legend(loc=4)
plt.show()
```

```python
a=data['facecream'].sum()
b=data['facewash'].sum()
c=data['toothpaste'].sum()
e=data['bathingsoap'].sum()
f=data['shampoo'].sum()
g=data['moisturizer'].sum()

class_product = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 
           'shampoo','moisturizer']
class_mean = [a,b,c,e,f,g]
explode = [0,0.03,0.1,0.2,0.3,0.4]

plt.pie(class_mean, labels = class_product,explode=explode,autopct="%0.2f%%")
plt.show()
```

```python
month = data['month_number']
 

a=data['facecream']
b=data['facewash']
c=data['toothpaste']
e=data['bathingsoap']
f=data['shampoo']
g=data['moisturizer']
 
# Stackplot with X, Y, colors value
plt.stackplot(month, a , b , c , e , f , g ,
              colors =['r', 'c','k','m','b','g'])
 
plt.xlabel('month')
plt.ylabel('No of units')
 
# Title of Graph
plt.title('Representation of month and units')
 
# Displaying Graph
plt.show()
```
